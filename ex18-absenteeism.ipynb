#### Dummies

```python
# which are the 28 reasons we have substituted with numbers here?
# the numbers in the column 'Reason for Absence' do not represent the actual reasons for absence but are unique numbers for each reason.
# --> the numbers represent categories that are equally meaningful
# --> it's a categorical numerical values
# add numeric meaning to our categorical nominal values to perform quantitative analysis
# use dummy variables : an explanatory binary variable that takes the value 1 only when a certain categorical effect is present, and 0 if the same effect is absent
# this way, we can be certain that an individual has been absent from work because of one and only one particular reason
reason_columns=pd.get_dummies(df['Reason for Absence'], drop_first= True,dtype=int)
reason_columns
```

```python
reason_columns['check']=reason_columns.sum(axis=1)
reason_columns
# create a new column 'check' that sums up the values in all the columns to make sure that each row has maximum one 1
# check=0 --> all 0s in a row (this means all the reasons for absence are absent for this) --> missing value
# check=1 --> one 1 in a row (this means that there's a single value in a row equals to 1) --> not missing value
# set the 'check' column to 1 (axis=1 means we are summing up the values along the horizontal axis)
```

```python
reason_columns['check'].sum(axis=0)
# now apply the sum method for the 'check' column directly --> add the values along the vertical axis
```

```python
reason_columns['check'].unique()
# to see how many different values we will retrieve from the 'check' column
# we get 1 value only (the number 1)
# this proofs that initially the reasons for absence column had no missing or incorrect values (check=1)
```

```python
reason_columns=reason_columns.drop(['check'], axis=1)
# now we can remove the 'check' column as it's no longer needed
```

##### now we should drop the reason 0 because we want to avoid potential multicollinearity issues

```python
reason_columns=pd.get_dummies(df['Reason for Absence'], drop_first=True,dtype=int)
reason_columns
```

#### Group the reasons for absence

```python
# get all the columns names
df.columns.values
```

```python
reason_columns.columns.values
# we get all the columns names for the reasons for absence (from 1 to 28)
```

```python
df=df.drop(['Reason for Absence'], axis=1)
df
# drop the 'Reason for Absence' column to avoid multicollinearity issues
# multicollinearity occurs when two or more independent variables are highly correlated, meaning that they are closely related and can provide the same information when used in a model.
# which is the case here because we've created 27 new columns (from 1 to 28 )
```

```python
# group these variables / classify them (from 1 to 28) --> re-organize them into groups 
# we will focus on qualitative analysis
# group1: from 1 to 14 (diseases)
# group2: from 15 to 17 (pregnancy)
# group3: from 18 to 21 (health issues-poisoning...)
# group4: from 22 to 28 (other reasons)
# each group itself will be a Dataframe object 
reason_columns.loc[:,1:14]
# we get 14 columns for group1 (diseases)
# 0: means that none of the values on the given row were equal to 1
# 1: somewhere among these 14 columns, there was a 1
```

```python
# we use the max method to get the maximum value in each row of the group1 (diseases)
reason_type_1=reason_columns.loc[:,1:14].max(axis=1)  
reason_type_2=reason_columns.loc[:,15:17].max(axis=1)
reason_type_3=reason_columns.loc[:,18:21].max(axis=1)
reason_type_4=reason_columns.loc[:,22:].max(axis=1)
reason_type_1
```

#### Concatenate Column Values: concatenate the 'Reason for Absence' columns with the 'Reason Type' columns 

```python
df=pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df
# reason_type_1 is 0 in this table, reason_type_2is 1, reason_type_3 is 2, reason_4 is 3
```
